```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import time
from splinter import Browser
import requests
```

```python
url = 'https://denver.craigslist.org/search/mca'
resp = requests.get(url)
soup = BeautifulSoup(resp.content, "html.parser")
```

```python

titles = soup.find_all("a", {"class": "result-title hdrlnk"})
titles_list = []

for i in range(120):
    titles_list.append(titles[i].get_text())
    
titles_list[:5]
```

```python
prices = soup.find_all("span", {"class": "result-price"})
prices_list = []

for i in range(0,120):
    prices_list.append(prices[i].get_text())
    
prices[0]
```

```python
df = pd.DataFrame(titles_list,prices_list)
df
```
