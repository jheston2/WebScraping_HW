```python
import requests
```

```python
data = requests.get('https://mars.nasa.gov/api/v1/news_items/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest')


data.json()['items']
```

```python
news_title = []

for i in data.json()['items']:
    news_title.append(i.get('title'))

news_title[:5]
```

```python
news_p = []

for i in data.json()['items']:
    news_p.append(i.get('description'))

news_p[:5]
```
